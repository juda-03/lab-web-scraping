```python
def scrape_books (min_rating, max_price):
    
    #Create the lists
    titles = []
    prices = []
    ratings = []
    availabilities = []
    genres = []
    upcs = []
    descriptions = []


    #we'll need the rating map
    rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
    }

    #loop through all pages

    BASE_URL = "http://books.toscrape.com/"

    for page in range (1,51):

        url = f"{BASE_URL}catalogue/page-{page}.html"

        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        books = soup.find_all("article", class_="product_pod")

        for book in books:

            #title
            title = book.find("h3").find("a").get("title")

            #price
            price_text = book.find("p", class_="price_color").text.strip()
            price = float(price_text.replace("£",""))

            #rating
            rating_tag = book.find("p", class_="star-rating")
            rating_word = rating_tag.get("class")[1]
            rating = rating_map[rating_word]

            #availability
            availability = book.find("p", class_="instock availability").text.strip()

            if rating >= min_rating and price <= max_price:

                #detailed_page
                detailed_page = book.find("h3").find("a").get("href")
                book_url = BASE_URL + "catalogue/" + detailed_page
                book_response = requests.get(book_url)
                book_soup = BeautifulSoup(book_response.content, "html.parser") 

                #genre
                breadcrumb_links = book_soup.find("ul", class_="breadcrumb").find_all("a")
                genre = breadcrumb_links[-1].text.strip()

                table_rows = book_soup.find("table").find_all("tr")
                upc = table_rows[0].find("td").text.strip()

                #description
                description_header = book_soup.find("div", id="product_description")
                description = description_header.find_next("p").text.strip()

                upcs.append(upc)
                titles.append(title)
                prices.append(price)
                ratings.append(rating)
                availabilities.append(availability)
                genres.append(genre)
                descriptions.append(description)


    df = pd.DataFrame({
            "UPC" : upcs,
            "Title": titles,
            "Price": prices,
            "Rating": ratings,
            "Availability": availabilities,
            "Genre": genres,
            "Description": descriptions,
            })
    return df


    


    
```
